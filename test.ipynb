```python
import threading
import uvicorn
from fastapi import FastAPI
from google.colab import output

# 2. FastAPIのアプリ（中身）を作る
app = FastAPI()

@app.get("/")
def read_root():
    return {"message": "これがFastAPIから送られたデータです！"}

# 3. サーバーを起動する関数
def run_server():
    # ポート 8000番 でサーバーを動かす
    uvicorn.run(app, host="127.0.0.1", port=8000)

# 4. 裏側（スレッド）でサーバーをスタートさせる
# (こうしないと、Colabのセルがここで止まってしまい、次の行に進めないため)
thread = threading.Thread(target=run_server)
thread.start()

# 5. Colabのポート8000番をブラウザで見れるように穴を開ける
output.serve_kernel_port_as_window(8000, path="/")
```
